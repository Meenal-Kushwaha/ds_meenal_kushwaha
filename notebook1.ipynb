```python
import pandas as pd

trades = pd.read_csv("/content/ds_meenal_kushwaha/csv_files/historical_data.csv")
sentiment = pd.read_csv("/content/ds_meenal_kushwaha/csv_files/fear_greed_index.csv")
```

```python
# Convert Date & Time

trades['Timestamp']=pd.to_datetime(trades['Timestamp'])
trades['Date']=trades['Timestamp'].dt.date

sentiment['Date']=pd.to_datetime(sentiment['date']).dt.date
```

```python
# Now we connect trades with sentiment using Date

merged=pd.merge(trades, sentiment, on='Date', how='inner')

merged=merged.dropna(subset=['classification'])
```

```python
# Win vs Loss Analysis

merged['result']=merged['Closed PnL'].apply(
    lambda x:'Win' if x>0 else 'Loss')

merged.groupby(['classification', 'result']).size()
```

```python
# Profit during Fear vs Greed

merged.groupby('classification')['Closed PnL'].mean()
```

```python
# Leverage Behavior

merged.groupby('classification')['Size USD'].mean()
```

# #Data Science Report

#Objective
The objective of this analysis is to study the relationship between trader behavior and market sentiment (Fear vs Greed). We aim to understand how profitability, risk, leverage, and trading activity change under different market sentiment conditions.

#Steps
  1. Loaded and cleaned the Bitcoin Fear & Greed Index dataset.
  2. Loaded and explored historical trader data from Hyperliquid.
  3. Merged trader data with market sentiment based on date.
  4. Performed exploratory data analysis (EDA) on profit, leverage, and trading volume.
  5. Compared trader performance during Fear vs Greed market conditions.
  6. Visualized key patterns and summarized insights.
#Key Insights
1. Traders tend to take higher leverage positions during Greed phases, indicating increased risk appetite.
2. Average profitability is more volatile during Greed periods compared to Fear periods.
3. During Fear sentiment, traders trade less frequently but show more controlled losses.
4. Certain traders consistently perform better during Fear markets, suggesting disciplined strategies outperform emotional trading.
#Conclusion
Market sentiment strongly influences trader risk-taking behavior.
